```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path='/content/Virat-Kohli-International-Cricket-Centuries.csv'

df=pd.read_csv(path)
df.head()
```

```python
df.dtypes
```

```python
df['Runs'] = df['Runs'].astype(str).str.replace('*', '')

df['Runs'] = df['Runs'].astype(int)
```

```python
df.to_csv('virat_century.csv', index=False)
```

```python
# Total runs scored by Virat Kohli
total_runs = df['Runs'].sum()
print("Virat Total runs",total_runs)

# Average runs per match
average_runs_per_match = df['Runs'].mean()
print("Virat Avg. run per match",average_runs_per_match)


# Maximum runs scored in a single match
max_runs = df['Runs'].max()
print('Virat Max Run',max_runs)


# Minimum runs scored in a match
min_runs = df['Runs'].min()
print('Virat Minimum Run',min_runs)
```

```python
# Runs distribution across different matches (histogram or box plot)
plt.figure(figsize=(10, 6))
sns.histplot(df['Runs'], bins=10, kde=True)
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.title('Runs Distribution')
plt.show()
```

```python
# Total runs scored against each opponent
total_runs_against_opponent = df.groupby('Against')['Runs'].sum()
print('Total runs against opponent',total_runs_against_opponent)
print('\n------------------------------------------')

# Average runs scored against each opponent
average_runs_against_opponent = df.groupby('Against')['Runs'].mean()
print('\n Avg. runs against opponent',average_runs_against_opponent)
```

```python
# Best and worst performances against specific opponents
best_performance = df[df.groupby('Against')['Runs'].transform('max') == df['Runs']]
best_performance
```

```python
worst_performance = df[df.groupby('Against')['Runs'].transform('min') == df['Runs']]
worst_performance
```

```python
# Performance analysis based on batting position
runs_by_position = df.groupby('Position')['Runs'].sum()
print('Runs by position',runs_by_position)

# Average runs scored at different batting positions
avg_runs_by_position = df.groupby('Position')['Runs'].mean()
print('\n Avarage run by position',avg_runs_by_position)
```
